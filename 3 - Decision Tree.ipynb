## What is Decision Tree?
A Decision Tree is a popular supervised learning algorithm used for both classification and regression tasks. It's a versatile model that can handle both categorical and numerical data and is known for its simplicity and interpretability. Decision Trees are particularly useful for data exploration and providing insights into the decision-making process.\

### How Decision Trees Work:

- **Tree Structure**:
  - A Decision Tree consists of nodes, branches, and leaves. Each node represents a feature/attribute, each branch represents a decision based on that feature, and each leaf node represents the outcome (class label or predicted value).
  
- **Decision-making Process**:
  - The tree structure is built recursively by splitting the dataset into subsets based on the values of input features. The splits are made at each node to maximize the information gain or minimize impurity.
  
- **Splitting Criteria**:
  - The choice of the feature and the threshold for splitting are determined based on metrics such as Gini impurity (for classification) or variance reduction (for regression).
  
- **Stopping Criteria**:
  - The tree-growing process stops when a certain stopping criterion is met, such as reaching a maximum depth, minimum number of samples in a leaf node, or when further splitting does not improve the model's performance.
  

## What are the problems a Decision Tree solves?

1. **Classification Trees**:
   - Used for classification tasks where the target variable is categorical. Each leaf node represents a class label, and the majority class in a leaf node is assigned as the predicted class.

2. **Regression Trees**:
   - Used for regression tasks where the target variable is continuous. Each leaf node represents a predicted value, typically the mean or median of the target variable in that node.

## Advantages of Decision Trees:

- **Interpretability**: Decision Trees provide a clear and easy-to-understand decision-making process, making them suitable for explaining the model's predictions.
- **Versatility**: Decision Trees can handle both numerical and categorical data without requiring feature scaling or one-hot encoding.
- **Non-parametric**: Decision Trees make fewer assumptions about the underlying data distribution compared to parametric models like linear regression.
- **Feature Importance**: Decision Trees can provide insights into feature importance by analyzing the splits and their impact on the model's performance.

## Disadvantage of Decision Trees:

- **Overfitting**: Decision Trees are prone to overfitting, especially when the tree is deep and complex. Techniques like pruning, limiting the tree depth, or using ensemble methods like Random Forests can help mitigate this issue.
- **Instability**: Small variations in the training data can lead to different tree structures, making Decision Trees less stable compared to other algorithms.
- **Bias towards Features with Many Levels**: Decision Trees tend to favor features with many levels (high cardinality), leading to biased splits.

## Use Case

```python
# import Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
file = "Datasets/Decision_Tree_ Dataset.csv"
balance_data = pd.read_csv(file)
print('Dataset:: \n')
balance_data.sample(5)
```

```python
# extracting the dependent and independent variables
x = balance_data.iloc[:, :-2].values
y = balance_data.iloc[:, 5].values
```

```python
# splitting the dataset into training set and test set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
# fitting Decision Tree to the Training set
clf_entropy = DecisionTreeClassifier() # creates an instance of Decision Tree
clf_entropy.fit(x_train, y_train)
```

```python
# Make Predictions
y_pred = clf_entropy.predict(x_test)
y_pred
```

```python
# Checking Accuracy
accuracy = accuracy_score(y_test,y_pred)
print('Accuracy Score is:', accuracy)
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Plotting the decision tree
plt.figure(figsize=(20,10))
plot_tree(clf_entropy, feature_names=balance_data.columns[:-2], filled=True)
# Add labels
plt.xlabel('Features', fontsize=14)
plt.ylabel('Decision Tree', fontsize=14)
plt.show()
```
